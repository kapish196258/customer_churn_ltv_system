```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/processed/feature_engineered_data.csv")
df.head(2)
```

```python
df.shape
```

**Observation:** The feature-engineered dataset contains 7,032 customer records and 31 numerical features ready for model preparation.

```python
# Separating Features and Target
X = df.drop("Churn", axis=1)
y = df["Churn"]
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

*The independent variables were stored in `X`, while the target variable (`Churn`) was stored in `y` for supervised machine learning.*

```python
# Train-Test Split
X_train, X_test, y_train,y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

*The dataset was divided into training and testing sets using an 80:20 split. Stratified sampling was applied to preserve the original churn distribution in both datasets.*

```python
# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print(X_train_scaled.shape)
print(X_test_scaled.shape)
```

*Standardization was applied to numerical features using `StandardScaler`. The scaler was fitted only on the training data and then applied to the test data to avoid data leakage.*

```python
# Save prepared data
pd.DataFrame(X_train_scaled).to_csv(
    "../data/processed/X_train_scaled.csv",
    index=False
)
pd.DataFrame(X_test_scaled).to_csv(
    "../data/processed/X_test_scaled.csv",
    index=False
)
y_train.to_csv(
    "../data/processed/y_train.csv",
    index=False
)
y_test.to_csv(
    "../data/processed/y_test.csv",
    index=False
)
print("Prepared datasets saved successfully!")
```

# Key Takeaways

* Prepared a machine-learning-ready dataset containing **7,032 customer records and 31 features**.
* Defined **features (`X`)** and the **target variable (`y`)** for churn prediction.
* Performed an **80:20 stratified train-test split** to preserve the original churn distribution.
* Applied **feature scaling using StandardScaler** to standardize numerical features.
* Prevented **data leakage** by fitting the scaler only on the training dataset.
* Generated the final **training and testing datasets** for churn model development and evaluation.
